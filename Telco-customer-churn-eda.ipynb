### Convert to Nnumeric and check for NaN values

```python
telco_df.drop('customerID', axis=1, inplace=True)
```

### Gender Distribution

```python
ax = (telco_df['gender'].value_counts()*100.0 /len(telco_df)).plot(kind='bar', rot=30,
                                                                   color = ['#4D3425','#FCA503'])
ax.set_ylabel('% of Customers')
ax.set_xlabel('Gender')
ax.set_title('Gender Distribution')

# Create a list to collect the plt.patches data
totals = []

# Find the values and append to list
for i in ax.patches:
    totals.append(i.get_width())

# Set individual bar lables using above list
total = sum(totals)

for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_x()+.15, i.get_height()-3.5, \
            str(round((i.get_height()/total), 1))+'%',
            fontsize=12,
            color='white', 
            weight = 'bold')
```

## Convert categorical values to boolean

```python
### Replace Gender {Male: 1, Female: 0}

telco_df['gender'] = telco_df['gender'].map({'Male': 1,'Female': 0})

### Replace Partner {Yes: 1, No: 0}

telco_df['Partner'] = telco_df['Partner'].map({'Yes': 1,'No': 0})

### Replace Dependents {Yes: 1, No: 0}

telco_df['Dependents'] = telco_df['Dependents'].map({'Yes': 1,'No': 0})

### Replace PhoneService {Yes: 1, No: 0}

telco_df['PhoneService'] = telco_df['PhoneService'].map({'Yes': 1,'No': 0})

### Replace Churn {Yes: 1, No: 0}

telco_df['Churn'] = telco_df['Churn'].map({'Yes': 1,'No': 0})

telco_df.head()
```

## Finding Outliers 

```python
continous_features = ["tenure", "MonthlyCharges", "TotalCharges"]
continous_telco_df = telco_df[continous_features]
continous_telco_df.describe()
```

```python
continous_telco_df.head()
```

## Calculate IQR

```python
Q1 = continous_telco_df.quantile(0.25)
Q3 = continous_telco_df.quantile(0.75)
IQR = Q3 - Q1

print("Q1")
print(Q1)
print("Q3")
print(Q3)
print("IQR")
print(IQR)
((continous_telco_df < (Q1 - 1.5 * IQR)) |(continous_telco_df > (Q3 + 1.5 * IQR))).any()
```

There are no outliers in the numerical data

### Boxplot with churn and tenure

Customers tenure ranges from 1 month to about 70 months. It is easy to see that the tenure of customers who don’t churn is longer than churn customers.

### Boxplot with churn and Total Charges

```python
sns.set_palette(sns.color_palette('Blues_d', 2))
sns.boxplot(x='Churn', y='TotalCharges', data=telco_df)
```

- For churned customers, they have a lower median point and a smaller range of IQR.
- It is reasonable to assume that since most of the churned customers did not stay with the company for a long time, thus they have a lower median total charge.
- It’s interesting to see that there are some outliners for churned customers. Their total charges are much higher than most of the customers.

### Partner vs Dependent

```python
fig,axis = plt.subplots(1,2,figsize=(12,6))
axis[0].set_title('Has partner')
axis[1].set_title('Has dependents')
axis_y = 'Percentage of customers'

telco_df_partner = telco_df.groupby('Partner')['Churn'].value_counts()/len(telco_df)
telco_df_partner = telco_df_partner.to_frame().rename({"Churn":axis_y}, axis=1).reset_index()
ax = sns.barplot(x='Partner', y=axis_y, hue='Churn', data=telco_df_partner, ax=axis[0])


telco_df_dependent = telco_df.groupby('Dependents')['Churn'].value_counts()/len(telco_df)
telco_df_dependent = telco_df_dependent.to_frame().rename({"Churn":axis_y}, axis=1).reset_index()
ax = sns.barplot(x='Dependents', y=axis_y, hue='Churn', data=telco_df_dependent, ax=axis[1])
```

## Transform Data with Label Encoder

```python
telco_df_discrete = telco_df.drop(labels = ['MonthlyCharges','TotalCharges'], axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
col = list(telco_df_discrete)
telco_df_discrete[col] = telco_df_discrete[col].apply(LabelEncoder().fit_transform)
telco_df_discrete.head()
```
